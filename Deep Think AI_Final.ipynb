```python
# converting it into a long format instead of operating it on wide format 
df_long = data.melt(
    id_vars=['Warehouse id', 'Region', 'SKU id'],  
    var_name='Month',                              
    value_name='Sales'
) 
```

```python
df_long.head()
```

```python
df_long['Month'] = pd.to_datetime(df_long['Month'], format='%b-%y') # converting to proper format 
```

```python
df_long.head()
```

```python
df_long.set_index('Month', inplace=True) # setting month as index 
```

```python
df_long[df_long['SKU id'] == 'SKU-1'].plot(figsize=(10, 6), title='Monthly Sales for SKU-1', xlabel='Month', ylabel='Sales')
plt.show()
```

```python
# one conclusion is that the sales of SKU-1 peaked during 2020-11 to 2020-12
```

```python
# Now taking the trends over time aggregating across all SKUs and warehouses
overall_sales = df_long.groupby('Month')['Sales'].sum()
overall_sales.plot(figsize=(10, 6), title='Overall Monthly Sales', xlabel='Month', ylabel='Total Sales')
plt.show()
```

```python
# regional analysis 
regional_sales = df_long.groupby('Region')['Sales'].sum()
regional_sales.plot(kind='bar', figsize=(8, 5), title='Total Sales by Region', ylabel='Total Sales')
plt.show()
```

```python
# warehouse specific analysis 
warehouse_sales = df_long.groupby(['Warehouse id', 'Month'])['Sales'].sum().unstack(level=0)
warehouse_sales.plot(figsize=(12, 6), title='Monthly Sales by Warehouse', xlabel='Month', ylabel='Total Sales')
plt.show()
```

```python
# The top 5 SKUs 
top_skus = df_long.groupby('SKU id')['Sales'].sum().sort_values(ascending=False).head(5)
print(top_skus)
top_skus.plot(kind='bar', figsize=(8, 5), title='Top 5 SKUs by Total Sales', ylabel='Total Sales')
plt.show()
```

```python
# seasonal trends 
df_long['MonthOnly'] = df_long.index.month
monthly_sales = df_long.groupby('MonthOnly')['Sales'].mean()
monthly_sales.plot(kind='line', figsize=(10, 6), title='Average Sales by Month', xlabel='Month', ylabel='Average Sales')
plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```

```python
# correlation analysis 
pivot_table = df_long.pivot_table(index='Month', columns='Region', values='Sales', aggfunc='sum')
print(pivot_table.corr())
```

```python
# Calculate IQR
Q1 = df_long['Sales'].quantile(0.25)
Q3 = df_long['Sales'].quantile(0.75)
IQR = Q3 - Q1

# Define the bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify and flag outliers
df_long['Outlier'] = (df_long['Sales'] < lower_bound) | (df_long['Sales'] > upper_bound)

# Replace outliers with NaN
df_long.loc[df_long['Outlier'], 'Sales'] = np.nan

# Fill missing values (imputed) using the median
df_long['Sales'] = df_long['Sales'].fillna(df_long['Sales'].median())

# (Optional) Verify the changes
print(df_long['Outlier'].sum(), "outliers were handled.")
```
